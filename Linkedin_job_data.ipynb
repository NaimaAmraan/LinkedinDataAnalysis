### Clean up columns using Regex (sequence of characters that forms a search pattern)

```python
#remove the text in no_of_employ containing strings and store results in the column 'extracted'

df['extracted'] = df['no_of_employ'].str.extract(r'(\d{1,3}(?:,\d{3})?-\d{1,3}(?:,\d{3})?)')

df.head()
```

```python
#reassign the column 'no_of_employ' to the column 'extracted' 

df['no_of_employ'] = df['extracted']

df.head()
```

```python
#remove the 'followers' text in linkedin followers so as to only have numbers. This regex pattern 
#matches numbers with commas (thousands separators) and uses word boundaries \b to ensure that we 
#capture entire numbers.

df['followers_extracted']= df['linkedin_followers'].str.extract(r'(\d{1,3}(?:,\d{3})*\b)')

df.head()
```

```python
#reassign column 'linkedin_followers' to 'followers_extracted'

df['linkedin_followers'] = df['followers_extracted']

#change linkedin followers to integers

df['linkedin_followers'] = pd.to_numeric(df['linkedin_followers'].str.replace(',', ''), errors='coerce', downcast='integer')


df.head()
```

```python
#clean up alumni column

df['alumni_extracted']= df['alumni'].str.extract(r'(\d+)')

df.head()
```

```python
#reassign the values in the 'alumni extracted' column to 'alumni' column
df['alumni'] = df['alumni_extracted']
df.head()
```

```python
#extract just the job title in 'job' column
df['title_extracted'] = df['job'].str.extract(r'(\b[a-zA-Z\s]+)\b')
df.tail()
```

```python
#assign the title extracted values back to the job column
df['job'] = df['title_extracted']
df.head()
```

```python
#look at the occurences of jobs and how many times they appear

df.groupby('job')['job'].count()
```

```python
#clean up 'hiring_person' column - convert all names to title case (previously some were all caps)
df['Hiring_person'] = df['Hiring_person'].str.title()
df.tail()
```

```python
#drop all the 'extracted' columns (extra that we do not need in the analysis)

df.drop(columns = ['followers_extracted'], inplace = True)
```

```python
#split minimum and maximum employees to get the numbers version of no_of_employ column


df[['Min_Employees', 'Max_Employees']] = df['no_of_employ'].str.split('-', expand=True)


df['Min_Employees'] = df['Min_Employees'].str.replace(',', '')
df['Max_Employees'] = df['Max_Employees'].str.replace(',', '')


df['Min_Employees'] = pd.to_numeric(df['Min_Employees'], errors = 'coerce')
df['Max_Employees'] = pd.to_numeric(df['Max_Employees'], errors = 'coerce')


```

*Let's check to see whether we'll find any correlation in the data*

```python
#find the numeric columns

numeric_columns = df.select_dtypes(include='number')


#build the correlation matrix

correlation_matrix = numeric_columns.corr()
print(correlation_matrix)
```

*Visualize the correlations using a correlation heatmap*

```python
#import visualization packages

import seaborn as sns

import matplotlib 

import matplotlib.pyplot as plt
```

```python
#plot the correlation heatmap

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlation Heatmap")
plt.show()
```
